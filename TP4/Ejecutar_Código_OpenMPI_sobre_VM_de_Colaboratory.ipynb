***Pasos a seguir para preparar la Máquina Virtual:***

1) Instalar Open MPI y todos los paquetes necesarios.

2) Montar google drive en la mv.

3) Posicionarse sobre la carpeta de google drive donde están los archivos escritos en C. Por defecto en esa carpeta se compilarán los archivos para su ejecución.

4) Compilar el código y ejecutarlo. 

*Nota:* la ejecución de los comandos Linux deben ser llamados precedidos por el simbolo !

```
# 2) Montar google drive en la mv.
from google.colab import drive
drive.mount('/content/drive/')
```
